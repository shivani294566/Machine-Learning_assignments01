```python
#  Prime Numbers
# A prime number is greater than 1 and divisible only by 1 and itself.

def is_prime(num):
    """Check if a number is prime"""
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):  # check divisors up to sqrt(num)
        if num % i == 0:
            return False
    return True

n = int(input("Enter a number to check if it's prime: "))
print(f"Is {n} a prime number? {is_prime(n)}")
```

```python
# Squares of Even/Odd Numbers
# Choice: Print squares of even numbers between 100 and 200

print("Squares of even numbers from 100 to 200:")
for i in range(100, 201):
    if i % 2 == 0:  # even check
        print(f"{i}^2 = {i**2}")
```

```python
# Word Counter
from collections import Counter

input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Split into words (basic split by space)
words = input_text.split()

# Count frequency
word_count = Counter(words)

print("Word Frequencies:")
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

```python
# Exercise 5: Palindrome Check
import re

def is_palindrome(s):
    """Return True if input string is a palindrome"""
    # Remove non-alphanumeric characters and convert to lowercase
    cleaned = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return cleaned == cleaned[::-1]

# Example usage
test_str = input("Enter a string to check if it's a palindrome: ")
print(f"Is '{test_str}' a palindrome? {is_palindrome(test_str)}")
```
